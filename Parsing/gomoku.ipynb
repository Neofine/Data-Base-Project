```python
from random import randint
import sys

# Losowa liczba całkowita od minimum do maximum zwrócona jako string, dopełniona prefixem zero jeśli jest jednocyfrowa.
def rand_string(minimum, maximum):
    rand = str(randint(minimum, maximum))
    if len(rand) < 2:
        rand = '0' + rand
    return rand

# Wygenerowana data of 2010 do 2021, godzina między 8 o 20. Ustaw delimiter jaki lubisz.
def rand_date_time():
    return rand_string(2018, 2020) + '-' + rand_string(1, 12) + '-' + rand_string(1, 28) + ' ' + rand_string(8, 19) + ':' + rand_string(0, 59) + ':' + rand_string(0, 59)

# Generujemy sobie number_of_dates dat.
number_of_dates = 250
list_of_dates = []
for i in range(number_of_dates):
    date = rand_date_time()
    list_of_dates.append(date)

# Sortujemy od najwcześniejszej do najnowszej. Jak wolisz odwrotnie to reverse zmień na True.
list_of_dates.sort(reverse = False)


original_stdout = sys.stdout

with open('datesList.txt', 'w') as f:
    sys.stdout = f
    for i in range(number_of_dates):
        print(list_of_dates[i])
    sys.stdout = original_stdout
```

```python
import sys
import random
import glob, os
os.chdir("/home/neofine/CLionProjects/BD")
ini = open("logins.txt", "r")

xd = ini.read()

split = xd.split('\n')

split.pop()

print(split)

def get_rand_pair(tab):
    people_amount = 30
    rand_from = []
    
    while(people_amount != 0):
        a = random.choice(tab)
        while(a in rand_from):
            a = random.choice(tab)
        rand_from.append(a)
        people_amount = people_amount - 1
    
    a = random.choice(rand_from)
    b = random.choice(rand_from)
    while(a == b):
        b = random.choice(rand_from)
    return (a, b)

games_generated = 250

original_stdout = sys.stdout

def count_lines():
    fp = open('Rozgrywka.sql', 'r')
    games = fp.read()
    spliterski = games.split("\n")
    return len(spliterski) - 1
    
prev_games = count_lines()
print(prev_games)

duel = []

for i in range(games_generated):
    pair = get_rand_pair(split)
    duel.append((pair[0], pair[1]));
        
def print_insert_duel(tab):
    global games_generated
    original_stdout = sys.stdout

    with open('Uczestnicy.sql', 'a') as f:
        sys.stdout = f

        for i in range(games_generated):
            print("INSERT INTO UCZESTNICY VALUES (", prev_games + i + 1, ", '", tab[i][0], "');", sep="");
            print("INSERT INTO UCZESTNICY VALUES (", prev_games + i + 1, ", '", tab[i][1], "');", sep="");

        sys.stdout = original_stdout
    
print_insert_duel(duel)
```

```python
import glob, os
os.chdir("/home/neofine/CLionProjects/BD/Standard")

all_games = []

def pop_last(line):
    split = line.split(",")
    return split[0] + "," + split[1]

def print_date_T(date):
    return "TIMESTAMP '" + date + "'"

i = 0

ansy = []

timestamp = ["2+1 15x15", "3+2 15x15", "5+0 15x15", "10+5 15x15", "10+0 15x15", "15+10 15x15", "30+0 15x15", "30+20 15x15",
             "2+1 20x20", "3+2 20x20", "5+0 20x20", "10+5 20x20", "10+0 20x20", "15+10 20x20", "30+0 20x20", "30+20 20x20"]
#timestamp = ["2+1", "3+2", "5+0", "10+5", "10+0", "15+10", "30+0", "30+20"]

def get_line(prev_games, i, duel, list_of_dates):
    return "INSERT INTO ROZGRYWKA VALUES (", prev_games + i + 1, ", 'GOMOKU', '", random.choice(duel[i]), "', ", print_date_T(list_of_dates[i]), ", ", random.randrange(1, 5),", '", random.choice(timestamp), "');"

for file in glob.glob("*.psq"):
    #print(file)
    ini = open(file, "r")
    game = ini.read()
    lines = game.split("\n")
    a_game = []
    for line in lines:
        if (len(line) > 0 and line[0].isdigit()):
            a_game.append(pop_last(line))

    ansy.append(get_line(prev_games, i, duel, list_of_dates))
    
    a_game.pop()
    all_games.append(a_game)
    
    if (i > games_generated - 2):
        break

    i = i + 1
    
    
os.chdir("/home/neofine/CLionProjects/BD")

with open('Rozgrywka.sql', 'a') as f:
    sys.stdout = f
    
    for line in ansy:
        print(line[0], line[1], line[2], line[3], line[4], line[5], line[6], line[7], line[8], line[9], line[10], sep="")
    
    sys.stdout = original_stdout
```
